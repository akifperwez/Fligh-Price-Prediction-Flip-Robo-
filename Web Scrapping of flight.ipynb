```python
# Importing Libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
#Opening the url
driver = webdriver.Chrome("chromedriver.exe")
url='https://www.yatra.com/'
driver.get(url)

time.sleep(2)
```

```python
search_button = driver.find_element_by_id('BE_flight_flsearch_btn')       # Locating search_button by xpath
search_button.click() 
```

```python
Airline=[]
Soruce=[]
Destination=[]
Departure_time=[]
Arrival_time=[]
Duration=[]
Total_stop=[]
Price=[]
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//span[@class = 'i-b text ellipsis']"):
        Airline.append(j.text)
Airline
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//p[@title = 'New Delhi']"):
        Soruce.append(j.text)
Soruce
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//p[@title = 'Mumbai']"):
        Destination.append(j.text)
Destination
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//div[@class = 'i-b pr']"):
        Departure_time.append(j.text)
Departure_time
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//p[@class = 'bold fs-15 mb-2 pr time']"):
        Arrival_time.append(j.text)
Arrival_time
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//p[@class = 'fs-12 bold du mb-2']"):
        Duration.append(j.text)
Duration
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//span[@class = 'dotted-borderbtm']"):
        Total_stop.append(j.text)
Total_stop
```

```python
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//div[@class = 'i-b tipsy fare-summary-tooltip fs-18']"):
        Price.append(j.text)
Price
```

```python
print(len(Airline),len(Soruce),len(Destination),len(Departure_time),len(Arrival_time),len(Duration),len(Total_stop),len(Price))
```

```python
Flight1=pd.DataFrame({})
Flight1["Airline"]=Airline[:180]
Flight1["Soruce"]=Soruce[:180]
Flight1["Destination"]=Destination[:180]
Flight1["Departure_time"]=Departure_time[:180]
Flight1["Arrival_time"]=Arrival_time[:180]
Flight1["Duration"]=Duration[:180]
Flight1["Total_stop"]=Total_stop[:180]
Flight1["Price"]=Price[:180]
Flight1[:180]
```
